Area Price Prediction Project
if area, bedroom increase then price increases
age of house increases then price decreases

```python
import pandas as pd
```

```python
df=pd.read_csv("multi.csv")
```

```python
df
```

```python
x=df[["area","bedrooms","age"]]
```

```python
x
```

```python
y=df.price
```

```python
y
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x['area'], x['bedrooms'], x['age'], c=y, cmap='viridis')
# Set labels
ax.set_title("Area, Bedrooms, Age vs Price")
ax.set_xlabel("Area")
ax.set_ylabel("Bedrooms")
ax.set_zlabel("Age")

# Add color bar to show price correlation
scatter = ax.scatter(x['area'], x['bedrooms'], x['age'], c=y, cmap='viridis')
colorbar = fig.colorbar(scatter, ax=ax, label='Price')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model.predict([[5000,3,5]])
```

```python
myarea = int(input("Enter the value of area: "))
mybedrooms = int(input("Enter the number of bedrooms: "))
myage = int(input("Enter the age of the house: "))

# Predict the price
myprediction = model.predict([[myarea, mybedrooms, myage]])
print("Predicted price is:", myprediction[0])
```
